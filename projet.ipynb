```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "/home/2inl2/davakov/dataScience/Tourism/datas/arrivees/englishArrivals.csv"
arrivals = pd.read_csv(path, skiprows=[0, 2], usecols=["Country Name", "2008", "2009", "2010", "2011", "2012", "2013"]);
pd.set_option('display.float_format', lambda x: '%.0f' % x);
```

```python
path = "/home/2inl2/davakov/dataScience/Tourism/datas/gains/englishGains.csv"
gains = pd.read_csv(path, skiprows=[0, 2], usecols=["Country Name", "2008", "2009", "2010", "2011", "2012", "2013"]);
pd.set_option('display.float_format', lambda x: '%.0f' % x);
```

```python
gains
```

```python
moneyPerPerson = (gains.set_index("Country Name") / arrivals.set_index("Country Name")).reset_index().dropna()
```
