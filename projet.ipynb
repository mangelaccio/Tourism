# Recette du tourisme par touriste (en dollars) par pays et par année

### Importation des bibliothèques

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Extraction des données concernant le nombre de touristes par pays et par années

```python
path = "datas/arrivees/englishArrivals.csv"
arrivals = pd.read_csv(path, skiprows=[0, 2], usecols=["Country Name", "2008", "2009", "2010", "2011", "2012", "2013"]);
pd.set_option('display.float_format', lambda x: '%.0f' % x);
```

```python
arrivals
```

### Extraction des données concernant les gains perçues par pays et par années en dollars

```python
path = "datas/gains/englishGains.csv"
gains = pd.read_csv(path, skiprows=[0, 2], usecols=["Country Name", "2008", "2009", "2010", "2011", "2012", "2013"]);
pd.set_option('display.float_format', lambda x: '%.0f' % x);
```

```python
gains
```

### Calcul de la recette par touriste (gains perçues du pays / nombre de touristes)
Suppression des pays dont les valeurs sont nulles ou absentes

```python
moneyPerPerson = (gains.set_index("Country Name") / arrivals.set_index("Country Name")).reset_index().dropna()
```

```python
moneyPerPerson
```

### Affichage graphique des données des pays de l'union européenne  

```python
countryList = ['Austria','Belgium','Bulgaria','Croatia','Cyprus','Czech Republic','Denmark','Estonia','France','Germany','Greece','Hungary','Ireland','Italy','Latvia','Lithuania','Luxembourg','Malta','Netherlands','Poland','Portugal','Romania','Slovenia','Spain','Sweden','United Kingdom']
my_plot = moneyPerPerson.set_index(['Country Name']).loc[countryList].plot(kind='bar', figsize=(20, 6))
plt.ylabel('Money per tourist',rotation=90)
```
